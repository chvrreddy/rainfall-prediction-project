```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

```

Load dataset and quick inspection

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/austin_weather.csv')
df
```

```python
#drop the unnecessary columns
df = df.drop(['Events', 'Date','SeaLevelPressureLowInches'], axis=1)
df.head()
```

```python
df = df.replace('T', 0.0)
df = df.replace('-', 0.0)
df['PrecipitationSumInches'] = pd.to_numeric(df['PrecipitationSumInches'])
df.to_csv('austin_final.csv')
```

```python
x = df.drop(['PrecipitationSumInches'],axis = 1)
y = df['PrecipitationSumInches']
```

```python
#reshape it into a 2d vector
y = y.values.reshape(-1,1)
```

```python
day_index = 798
days = [i for i in range(y.size)]
```

```python
#intilize the linear regression classifier
clf = LinearRegression()
clf.fit(x,y)
```

```python
#plot a graph
print('The precipitation trend graph: ')
plt.scatter(days,y,color='b')
plt.scatter(days[day_index],y[day_index],color='r')
plt.title('Precipitation level')
plt.xlabel('Days')
plt.ylabel('Precipitation in inches')
plt.show()
x_vis = x.filter(['TempAvgF','DewPointAvgf','HumidityAvgPercent','SeaLevelPressureAvgInches','VisibilityAvgMiles','WindAvgMPH'])
```

```python
print("THe precipitation vs attributes trend graph : ")

# Explicitly convert columns to numeric to ensure clean data for plotting
x_vis_clean = x_vis.apply(pd.to_numeric, errors='coerce')

for i in range(x_vis_clean.columns.size):
  plt.subplot(3,2,i+1)
  plt.scatter(days,x_vis_clean[x_vis_clean.columns.values[i]],color = 'g')
  plt.scatter(days[day_index],x_vis_clean[x_vis_clean.columns.values[i]].iloc[day_index],color='r')
  plt.title(x_vis_clean.columns.values[i])

plt.tight_layout() # Added to prevent overlapping subplots
plt.show()
```
